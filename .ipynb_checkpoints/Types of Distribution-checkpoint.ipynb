```python
from scipy.stats import norm as gaussian
from scipy.stats import multivariate_normal as gaussian_multi
from scipy.stats import bernoulli as bern
from scipy.stats import multinomial
from scipy.stats import expon
from scipy.stats import laplace
from scipy.stats import binom
from scipy.stats import poisson
import numpy as np
import matplotlib.pylab as plt
```

```python
# Normal Distribution
x = np.linspace(-6,6,1000)
mu, sigma = 0, 2
plt.plot(x, gaussian.cdf(x,mu,sigma));
plt.plot(x, gaussian.pdf(x,mu,sigma));
```

```python
# Multivariate Normal Distribution
x = np.linspace(-6, 6, 1000)
mu, sigma = 0,0.5 #Note: sigma in this case is the covariance matrix
plt.plot(x, gaussian_multi.cdf(x,mu,sigma));
plt.plot(x, gaussian_multi.pdf(x,mu,sigma));
```

```python
# Exponential Distribution
x = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100)
plt.plot(x, expon.pdf(x),'r-', lw=5, alpha=0.6, label='expon pdf');
plt.plot(x, expon.cdf(x), 'b-', lw=5, alpha=0.5, label='expon cdf');
plt.legend(loc='best', frameon=False)
plt.show()
```

```python
# Laplace Distribution
x = np.linspace(laplace.ppf(0.01),laplace.ppf(0.99), 100)
plt.plot(x, laplace.pdf(x),'r-', lw=5, alpha=0.6, label='laplace pdf');
plt.plot(x, laplace.cdf(x),'b-', lw=5, alpha=0.5, label='laplace cdf');
plt.legend(loc='best', frameon=False) #Note: the purple is the overlap
plt.show()
```

```python
# Multinomial Distribution
rv = multinomial(8, [0.3, 0.2, 0.5])
rv.pmf([1, 3, 4])
```

```python
# Bernoulli Distribution
p = 0.3
x = np.arange(bern.ppf(0.01, p),(bern.ppf(0.99,p))) # Note: ppf finds the value associated with the 
                                                    # given probability
plt.plot(x, bern.pmf(x,p),'bo',ms=8,label='bernoulli pmf');
plt.vlines(x,0,bern.pmf(x,p), colors='b', lw=5, alpha=0.5);
#plot(x, bern.cdf(x,p));
plt.legend(loc='best', frameon=False)
plt.show()
```

```python
# Bionomial Distribution
n, p = 5, 0.4
x = np.arange(binom.ppf(0.01, n, p),binom.ppf(0.99, n, p))
plt.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
plt.plot(x, binom.cdf(x, n, p), 'ro', ms=8, label='binom cdf')
plt.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
plt.legend(loc='best', frameon=False)
plt.show()
```

```python
# Poisson Distrubtion
mu = 0.6
x = np.arange(poisson.ppf(0.01, mu),poisson.ppf(0.99, mu))
plt.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
plt.plot(x, poisson.cdf(x, mu), 'ro', ms=8, label='poisson cdf')
plt.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
plt.legend(loc='best', frameon=False)
plt.show()
```
